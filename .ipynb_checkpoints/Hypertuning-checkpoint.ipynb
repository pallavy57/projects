```python
#checking outliers
outliers = []
def detect_outliers_iqr(df):
    data = sorted(df)
    q1 = np.percentile(df, 25)
    q3 = np.percentile(df, 75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers# Driver code
columns_outliers = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness","Insulin",'DiabetesPedigreeFunction','Age']
sample_outliers_ai = detect_outliers_iqr(df["Pregnancies"])
print("Outliers from IQR method: ", sample_outliers_ai)

print("////////////////////////////////////////////////")

sample_outliers_ac = detect_outliers_iqr(df["Glucose"])
print("Outliers from IQR method: ", sample_outliers_ac)
print("////////////////////////////////////////////////")
sample_outliers_al = detect_outliers_iqr(df["BloodPressure"])
print("Outliers from IQR method: ", sample_outliers_al)
print("////////////////////////////////////////////////")
sample_outliers_at = detect_outliers_iqr(df["SkinThickness"])
print("Outliers from IQR method: ", sample_outliers_at)
print("////////////////////////////////////////////////")
sample_outliers_ac = detect_outliers_iqr(df["Insulin"])
print("Outliers from IQR method: ", sample_outliers_ac)
print("////////////////////////////////////////////////")
sample_outliers_at = detect_outliers_iqr(df["DiabetesPedigreeFunction"])
print("Outliers from IQR method: ", sample_outliers_at)
print("////////////////////////////////////////////////")
sample_outliers_ac = detect_outliers_iqr(df["Age"])
print("Outliers from IQR method: ", sample_outliers_ac)
```

```python
columns_outliers = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness","Insulin",'DiabetesPedigreeFunction','Age']

df['Pregnancies'] = np.where(df['Pregnancies'] > df['Pregnancies'].quantile(0.95), df['Pregnancies'].quantile(0.50), df['Pregnancies'])
df['BloodPressure'] = np.where(df['BloodPressure'] > df['BloodPressure'].quantile(0.95), df['BloodPressure'].quantile(0.50), df['BloodPressure'])
df['Glucose'] = np.where(df['Glucose'] > df['Glucose'].quantile(0.95), df['Glucose'].quantile(0.50), df['Glucose'])
df['SkinThickness'] = np.where(df['SkinThickness'] > df['SkinThickness'].quantile(0.95), df['SkinThickness'].quantile(0.50), df['SkinThickness'])
df['Insulin'] = np.where(df['Insulin'] > df['Insulin'].quantile(0.95), df['Insulin'].quantile(0.50), df['Insulin'])
df['DiabetesPedigreeFunction'] = np.where(df['DiabetesPedigreeFunction'] > df['DiabetesPedigreeFunction'].quantile(0.95), df['DiabetesPedigreeFunction'].quantile(0.50), df['DiabetesPedigreeFunction'])
df['Age'] = np.where(df['Age'] > df['Age'].quantile(0.95), df['Age'].quantile(0.50), df['Age'])
```

```python
df['Pregnancies'] = df['Pregnancies'].fillna(df['Pregnancies'].median())
df['BloodPressure'] = df['BloodPressure'].fillna(df['BloodPressure'].median())
df['DiabetesPedigreeFunction'] = df['DiabetesPedigreeFunction'].fillna(df['DiabetesPedigreeFunction'].mode())
df['Glucose'] = df['Glucose'].fillna(df['Glucose'].median())
df['SkinThickness'] = df['SkinThickness'].fillna(df['SkinThickness'].median())
df['Insulin'] = df['Insulin'].fillna(df['Insulin'].median())
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
#getting the upper limit and lowelimit and new
def returnNewShape(column):
        # Finding the IQR
        percentile25 = column.quantile(0.25)
        percentile75 =column.quantile(0.75)
        sort_data = np.sort(column)
        sort_data


        Q1 = np.percentile(column, 25, interpolation = 'midpoint') 
        Q2 = np.percentile(column, 50, interpolation = 'midpoint') 
        Q3 = np.percentile(column, 75, interpolation = 'midpoint') 

        print('Q1 25 percentile of the given data is, ', Q1)
        print('Q1 50 percentile of the given data is, ', Q2)
        print('Q1 75 percentile of the given data is, ', Q3)

        iqr = Q3 - Q1 
        print('Interquartile range is', iqr)

        # Finding the upper and lower limits
        upper_limit = percentile75 + 1.5 * iqr
        lower_limit = percentile25 - 1.5 * iqr


        # Trimming outliers
        new_df = df[column < upper_limit]
        return new_df,upper_limit,lower_limit
        Insulin
new_df = returnNewShape(df["DiabetesPedigreeFunction"]) 
new_df


        
```

```python
y = new_df[0]['Outcome'] 
X = new_df[0].drop(['Outcome'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
```

```python
knn_clf.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# We first create a KNN classifier instance ]
# prepare a range of values of hyperparameter K from 1 to 31 
# that will be used by GridSearchCV to find the best value of K.
```

```python
from sklearn.model_selection import GridSearchCV
k_range = list(range(1, 31)) 
param_grid = dict(n_neighbors=k_range)
# defining parameter range
grid = GridSearchCV(knn_clf, param_grid, cv=10, scoring='accuracy', return_train_score=False,verbose=1)
  
# fitting the model for grid search
grid_search=grid.fit(X_train, y_train)

print(grid_search.best_params_)
```

```python
accuracy = grid_search.best_score_ *100
print("Accuracy for our training dataset with tuning is : {:.2f}%".format(accuracy) )
```

```python
# this can be used to fit a KNN model and check its accuracy on the unseen test dataset.
knn = KNeighborsClassifier(n_neighbors=17)
# X_train, X_test, y_train, y_test
knn.fit(X_train, y_train)

y_test_hat=knn.predict(X_test) 

test_accuracy=accuracy_score(y_test,y_test_hat)*100

print("Accuracy for our testing dataset with tuning is : {:.2f}%".format(test_accuracy) )
```
